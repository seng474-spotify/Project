```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, LabelEncoder

def preprocess_data(file_path):
    # Load dataset
    df = pd.read_csv(file_path)

    
    topEightGenres = df["genre"].value_counts().index[:8]
    df = df[df["genre"].isin(topEightGenres)]
    
    # Drop 'genre' column from X and keep it for Y
    X = df.drop(columns=['Unnamed: 0', 'genre', 'artist_name', 'track_name', 'track_id'])
    Y = df['genre']
    
    scaler = MinMaxScaler()
    X = scaler.fit_transform(X)

    # Label encode the genre column
    label_encoder = LabelEncoder()
    Y = label_encoder.fit_transform(Y)

    return X, Y

# File path
file_path = "spotify_data.csv"

# Preprocess data
X, Y = preprocess_data(file_path)

print(X.shape)  
print(Y.shape)
```
